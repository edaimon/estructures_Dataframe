D'aquí deduim que la relació de 'ActualElapsedTime' es pot explicar perquè els vols 'Cancelled' i 'Diverted' no tenen temps real de vol. Per comprovar-ho, treiem els vols 'Cancelled' i 'Diverted' primer i veiem el resultat.

```python
df2 = df1 #fem un nou dataframe per no perdre les dades en cas d'error
df2
```

```python
# Creació de variable per guardar els vols cancel·lats
vols_cancellats = df2[df2['Cancelled']==1]
len(vols_cancellats)
```

Veiem que efectivament s'han restat 68 vols en totes les columnes. Per tant, els valors nuls d'altre columnes venen donants per 'Cancelled´. Fem lo mateix amb 'Diverted':

```python
# Creació de variable per guardar els vols desviats
vols_desviats = df2[df2['Diverted']==1]
print(len(vols_desviats))
```

798 és justament els vols amb valors nulls d'altres columnes.

```python
# Extracció dels registres on el vol està desviat
df2 = df2[df2['Diverted']!=1]
df2
```

```python
# Comprovem quins valors nuls queden al dataframe
df2.isnull().sum()
```

Aquí veiem que amb les columnes 'Cancelled' i 'Diverted' hem tret tots els valors nuls del dataset. Continuem eliminant la columna 'CancellationCode', que s'ha quedat buida de sentit

```python
# Treure columnca 'CancellationCode'
df2 = df2.drop('CancellationCode', axis=1)
print(df2)
```

#### Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...). 

```python
df3 = df2 #Nou df per no perdre dades anteriors

# Creació de nova columna amb la velocitat mitjana en milles per hora, dividint la distància pels temps real de vol i multiplicant per 60
df3['AvgSpeed'] =  (df3['Distance'] / df3['ActualElapsedTime']) * 60
```

```python
df3
```

```python
# Arrodonir els valors a dos decimals
df3['AvgSpeed'] = df3['AvgSpeed'].round(2)
df3
```

```python
# Per saber quins vols han arribat tard en una nova columna amb booleans (1 = Yes, 0 = No)
df3['DelayedFlight'] = np.where(df3['ArrDelay'] > 0, '1', '0')
df3
```

#### Fes una taula de les aerolínies amb més endarreriments acumulats.

```python
unique_carrier_delays = pd.DataFrame(df3.groupby('UniqueCarrier')['ArrDelay'].sum().sort_values(ascending=False))
unique_carrier_delays.head(5)
```

El top 5 d'aerolínies amb més retards són les següents:<br/>
WN - Southwest Airlines<br/>
AA - American Airlines<br/>
UA - United Airlines<br/>
MQ - Envoy Air (American Eagle)<br/>
OO - SkyWest Airlines<br/>


####  Quins són els vols més llargs?   

```python
# Els més llargs de distància
longest_flights = df3[['FlightNum', 'Distance']].sort_values(by='Distance', ascending=False)
longest_flights.head(5)
```

```python
# Els de més duració. Prenc 'ActualElapsedTime' com referència
timest_flights = df3[['FlightNum', 'AirTime', 'ActualElapsedTime']].sort_values(by='ActualElapsedTime', ascending=False)
timest_flights.head(5)
```

#### I els més endarrerits?

```python
# Els de més duració. Prenc 'ActualElapsedTime' com referència
timest_flights = df3[['FlightNum', 'AirTime', 'ActualElapsedTime']].sort_values(by='ActualElapsedTime', ascending=True)
timest_flights.head(5)
```

#### Busca les rutes més llargues

```python
# Els més llargs de distància, incloent-hi 'Origin' => 'Dest' & 'Dest' => 'Origin' com a mateixa ruta
longest_routes = df3[['Origin', 'Dest', 'Distance']].sort_values(by='Distance', ascending=False)
longest_routes = longest_routes.drop_duplicates(subset=['Distance']) # Eliminar rutes Origen-Destí duplicades o amb la mateixa distància
longest_routes.head(5)
```

#### i les que acumulen més retards.

```python
# Els més llargs de distància
timest_routes = pd.DataFrame(df3.groupby(['Origin', 'Dest'])['DelayedFlight'].sum())
timest_routes = timest_routes.sort_values(by='DelayedFlight', ascending=False)

timest_routes.head(5)
```

```python
# Com que concatena els 1, comprovem el tipus de dada que és 'DelayedFlight', ja que sembla que no és integer
print(df3['DelayedFlight'].dtypes)
```

```python
# Convertim la columna a integer
df3['DelayedFlight'] = df3['DelayedFlight'].astype(int)
```

```python
# Repetim l'operació per trobar les rutes amb més retards
timest_routes = pd.DataFrame(df3.groupby(['Origin', 'Dest'])['DelayedFlight'].sum())
timest_routes = timest_routes.sort_values(by='DelayedFlight', ascending=False)

timest_routes.head(5)
```
